```python
import pandas as pd
```

```python
teams =[
  "diamondbacks",
  "braves",
  "orioles",
  "redsox",
  "cubs",
  "whitesox",
  "reds",
  "guardians",
  "rockies",
  "tigers",
  "astros",
  "royals",
  "angels",
  "dodgers",
  "marlins",
  "brewers",
  "twins",
  "mets",
  "yankees",
  "athletics",
  "phillies",
  "pirates",
  "padres",
  "giants",
  "mariners",
  "cardinals",
  "rays",
  "rangers",
  "bluejays",
  "nationals"
]
```

```python
dataframes = {}

for i in teams:
    filepath = f'source/{i}.csv'
    dataframes[i] = pd.read_csv(filepath)
    dataframes[i]['team'] = i
    dataframes[i].info()
```

```python
dataframes['braves'].head()
```

```python
df = pd.concat(dataframes, ignore_index=True)
df.info()
```

```python
df.dropna(subset=['Player'], inplace= True)
df['Player'].info()
```

```python
df = df.drop(columns=['Rk','H▼','Player-additional', 'Pos.1']) 
df.info()
```

```python
df['Last_Char'] = df['Player'].str[-1]
df['dominant_hand'] = ''  

for idx, row in df.iterrows(): 
    if row['Last_Char'] == '*':    
        df.loc[idx, 'dominant_hand'] = 'Left'
    elif row['Last_Char'] == '#':
        df.loc[idx, 'dominant_hand'] = 'Both'
    elif row['Last_Char'] == '?':
        df.loc[idx, 'dominant_hand'] = 'Unknown'
    else:
        df.loc[idx, 'dominant_hand'] = 'Right'

```

```python
df.tail()
```

```python
# Remove suffix characters from Players column
for idx, row in df.iterrows():
    if row['Last_Char'] in ['*', '#', '?']:
        df.loc[idx, 'Player'] = row['Player'][:-1]

df["Player"].head(10)
```

```python
df = df.drop(columns=['Last_Char']) 
df.info()
```

```python
df = df[['Player', 'team', 'Age', 'Pos', 'dominant_hand','G', 'PA', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'SB', 'CS','BB', 'SO', 'HBP', 'TB', 'GIDP', 'SH', 'SF', 'IBB','BA', 'OBP', 'SLG', 'OPS','OPS+', 'rOBA', 'Rbat+', 'WAR','Awards']]
df.head()
```

```python
df.to_csv('api_teams_data.csv', index=False)
```
